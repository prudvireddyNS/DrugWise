```python
import pandas as pd
import json
```

```python
data = pd.read_csv('data/medicine_data.csv')
data.head()
```

```python
data.drop(columns=['product_price'], inplace=True)
```

```python
drug_interactions = [json.loads(entry) for entry in data['drug_interactions']]
drug_interactions = pd.DataFrame(drug_interactions)
```

```python
drug_interactions.head()
```

```python
drug_interactions.shape
```

```python
data.shape
```

```python
merged = pd.concat([data, drug_interactions], axis=1)
merged.drop(columns=['drug_interactions'], inplace=True)
```

```python
merged.head()
```

```python
# Filter rows where drug_interactions is not an empty list
df_filtered = merged[merged['drug'].apply(lambda x: len(x) > 0)]
df_filtered = merged[merged['brand'].apply(lambda x: len(x) > 0)]
df_filtered = merged[merged['effect'].apply(lambda x: len(x) > 0)]

# Reset the index after filtering
df_filtered.reset_index(drop=True, inplace=True)

# Print the description of the filtered DataFrame
print(df_filtered.head())
```
